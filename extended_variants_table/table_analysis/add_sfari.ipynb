```python
import pandas as pd
```

```python
variants_path = '/private10/Projects/Nave_Oded_Project/extended_variants_table/Results/processed_varicarta_table_2nd_run.csv'
variants_df = pd.read_csv(variants_path, low_memory=False)
variants_df.head()
```

```python
sfari_path = '/private10/Projects/Nave_Oded_Project/extended_variants_table/table_analysis/SFARI-Gene_genes_04-03-2025release_06-30-2025export.csv'
sfari_df = pd.read_csv(sfari_path, low_memory=False)
sfari_df.head()
```

```python
merged_df = variants_df.merge(sfari_df[['gene-symbol', 'gene-score', 'syndromic', 'ensembl-id', 'number-of-reports']], how='left', left_on='SYMBOL', right_on='gene-symbol')
merged_df.shape
```

```python
merged_df.drop(columns=['gene-symbol', 'ensembl-id'], inplace=True)
merged_df.rename(columns={'gene-score': 'sfari_gene_score', 'syndromic': 'sfari_syndromic', 'number-of-reports': 'sfari_number_of_reports'}, inplace=True)
merged_df.head()
```

```python
output_path = '/private10/Projects/Nave_Oded_Project/extended_variants_table/Results/processed_varicarta_table_2nd_run_with_sfari.csv'
merged_df.to_csv(output_path, index=False)
```
